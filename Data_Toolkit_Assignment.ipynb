### **1. What is NumPy, and why is it widely used in Python?**

**NumPy** (Numerical Python) is a library in Python used for numerical and scientific computing. It provides support for arrays (multidimensional matrices), mathematical functions, and efficient operations on large datasets.

**Why is it widely used?**
- **Efficient computation:** NumPy operations are implemented in C and optimized for performance, making them faster than Python loops or lists.
- **Multi-dimensional arrays:** Supports n-dimensional arrays, useful for scientific computations.
- **Broadcasting:** Simplifies operations on arrays of different shapes.
- **Integration:** Works seamlessly with other Python libraries like Pandas, Matplotlib, and Scikit-learn.

**Example:**
```python
import numpy as np

# Creating a NumPy array
arr = np.array([1, 2, 3, 4, 5])
print(arr * 2)  # Element-wise multiplication: [2, 4, 6, 8, 10]
```

### **2. How does broadcasting work in NumPy?**

**Broadcasting** allows NumPy to perform operations on arrays of different shapes without explicitly resizing them. It expands the smaller array along the dimensions to match the larger array.

**Rules for Broadcasting:**
1. The arrays must have compatible shapes.
2. Dimensions are compatible if they are the same or one of them is 1.

**Example:**
```python
import numpy as np

a = np.array([1, 2, 3])          # Shape: (3,)
b = np.array([[10], [20], [30]]) # Shape: (3, 1)

result = a + b  # Broadcasting: Shape becomes (3, 3)
print(result)
# Output:
# [[11 12 13]
#  [21 22 23]
#  [31 32 33]]
```

### **4. Explain the use of the `groupby()` method in Pandas.**

The `groupby()` method is used to split the data into groups based on some criteria, perform operations on each group, and combine the results.

**Example:**
```python
import pandas as pd

data = {
    'Department': ['HR', 'Finance', 'HR', 'IT', 'Finance'],
    'Employee': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
    'Salary': [5000, 6000, 5500, 7000, 6500]
}
df = pd.DataFrame(data)

# Grouping by 'Department' and calculating the average salary
avg_salary = df.groupby('Department')['Salary'].mean()
print(avg_salary)
```
**Output:**
```
Department
Finance    6250.0
HR         5250.0
IT         7000.0
Name: Salary, dtype: float64
```

### **6. What are the differences between NumPy arrays and Python lists?**

| Feature          | NumPy Arrays                       | Python Lists                |
|-------------------|------------------------------------|-----------------------------|
| **Data type**     | Homogeneous (all elements of the same type). | Heterogeneous (elements can have different types). |
| **Performance**   | Faster due to optimized C backend. | Slower due to dynamic typing. |
| **Size**          | Fixed-size.                        | Dynamic resizing.           |
| **Functionality** | Supports vectorized operations.    | Requires manual looping for element-wise operations. |

**Example:**
```python
# Python List
lst = [1, 2, 3, 4]
print([x * 2 for x in lst])  # Manual iteration: [2, 4, 6, 8]

# NumPy Array
import numpy as np
arr = np.array([1, 2, 3, 4])
print(arr * 2)  # Vectorized: [2, 4, 6, 8]
```

### **10. What is the significance of hierarchical indexing in Pandas?**

### **13. Why is handling missing data important in Pandas?**

Missing data can lead to incorrect analysis and skewed results. Properly handling missing data ensures the integrity of data analysis and modeling.

#### **Methods to Handle Missing Data**:
1. **Drop Missing Values**:
   ```python
   df.dropna(inplace=True)
   ```
2. **Fill Missing Values**:
   ```python
   df['A'].fillna(df['A'].mean(), inplace=True)
   ```
3. **Interpolation**:
   ```python
   df.interpolate(method='linear', inplace=True)
   ```

#### **Example**:
```python
import pandas as pd

data = {'A': [1, 2, None, 4], 'B': [None, 2, 3, 4]}
df = pd.DataFrame(data)

# Fill missing values with mean
df['A'] = df['A'].fillna(df['A'].mean())
print(df)
```

### **17. Explain the difference between `apply()` and `map()` in Pandas.**

- **`apply()`**: Applies a function to each row/column of a DataFrame or Series.
- **`map()`**: Applies a function element-wise to a Series.

#### **Example**:
```python
import pandas as pd

df = pd.DataFrame({'A': [1, 2, 3], 'B': [4, 5, 6]})

# Apply to rows
df['C'] = df['A'].apply(lambda x: x**2)

# Map to a Series
df['D'] = df['B'].map(lambda x: x*2)
print(df)
```

### **19. How does Pandas simplify time series analysis?**

Pandas provides:
- **DatetimeIndex**: Simplifies indexing and slicing by dates.
- **Resampling**: Aggregates data at different time intervals.
- **Rolling**: Computes moving averages.

#### **Example**:
```python
import pandas as pd

# Create a time series
dates = pd.date_range('20230101', periods=6)
data = pd.Series([1, 2, 3, 4, 5, 6], index=dates)

# Resample to weekly mean
print(data.resample('W').mean())
```
